## 2. Metadata Creation  
__Custom Functions__  
jwrangle.MQ_writeMetadata( )  
  
Metadata tabulates the test conditions for ALL experiments that shared the same MQ search and thuss all experiments that comprise the MQ outputs. Metadata can also be done in a spreadsheet program. Here, I have created my metadata programmatically instead (I simply find this easier).   
The metadata table gives users the opportunity to rename samples and define the experimental parameters for the data. This task can be expecially complex for MaxQuant because a unified output is generated even if distinctly separate experiments are searched as a batch and with different parameters applied.   
The function __jwrangle.MQ_writeMetadata( )__ will take a metadata table, rename all samples in the proteinGroups and evidence files, assign alternative filenames, and save new copies to be used in future analyses.

```python
#### Inspect column names
colnames = list(pGroups.columns.values)

#### Derive experiment names as a list
experiment_names = []
for i in colnames:
    if 'Intensity ' in i:
        experiment_names.append(i.replace('Intensity ', ''))

#### Fix up some of the names
mockexpt_names = []
for i in experiment_names:
    if i == 's48_rRNA1of10_Slot1-54_1_4211':
        i = 'p3833_48_PCprobe_A_Slot1-12_1_4223'
    elif i == 's49_rRNA1of10_Slot1-53_1_4210':
        i = 'p3833_49_PCprobe_B_Slot1-12_1_4223'
    i = i.replace('PC','254-')
    i = i.replace('NC', 'nCL-')
    mockexpt_names.append(i)

mockexpt_names   

#### Create a list of associated conditions
condition =  []
for i in mockexpt_names:
    condition.append(i.split('_')[2])

#### Create a list of associated replicate identifiers
replicate = []
for i in mockexpt_names:
    replicate.append(i.split('_')[3])
    
#### Create a more reader friendly list of sample names
samples =  []
for n, l in enumerate(condition):
    samples.append(l + '_' + replicate[n])

#### The MQ_groups dictionary assigns each condition to its MQ group parameter
MQ_groups = {'bead':['nCL-bead','254-bead'], 
             'probe':['nCL-probe','254-probe']}

#### We then use the condition list to create an MQ_groups list
Grp_parameters = []
for label in condition:
    for key, value in MQ_groups.items():
        if label in value:
            Grp_parameters.append(key)

expt_df = pd.DataFrame(
    {'experiment': experiment_names,
     'condition': condition,
     'replicate': replicate,
     'sample':samples,
     'measure':['Intensity']*len(samples),                  # adding this column allows our metadata file to be compatible with Proteus
     'MQgroups':Grp_parameters
    })

expt_df
```

```python
#### Write the modified proteins Groups and evidence files to a local folder
MQ_expt295 = jwrangle.MQ_writeMetadata(pGroups, evidence, expt_df, 'e298', cwd)
```

## 3. Re-Annotate the MaxQuant pGroups with stable Gene IDs  
__Functions__  
jweb.mapAnyID( )  
jwrangle.importMixedFiles( )  
  
MaxQuant does a good job of assigning a Gene name to each protein group. Presumably these gene names come from the FASTA. However:    
* Sometimes it fails to find a gene name  
* Sometimes it will assign an ID that is not a gene or include out-of-place characters    
* It doesn't always seem to be consistent  
* If the gene name originates from the FASTA then repeating the MQ search with an updated FASTA is the only way to update the gene IDs.  
* Use of a mapping service will standardise the ID conversion practices between my datasets and those of others, including RNA-Seq.  
  
To avoid these problems we will remap the Majority protein IDs to ENTREZ gene IDs.  ___jweb.mapAnyID( )___ will retrieve all possible genes for each protein group, and will also select a primary ID to singularly represent the group by a consistent method. This is a very flexible function, see help( ) for further explanation. From this point, the MQ 'Gene names' column will no longer be necessary.  This function can also handle ID mapping to and from almost any convention.  
  
Ensuring our proteins have a consistent gene naming strategy is essential for inter-experiment comparison and the later use of set methods. It also creates a standard that can be applied for accurately mapping RNA-Seq results and thus aid in future mapping of protein-RNA partners.

## 7. Drop Gene Duplicates and Filter Intensities by LFQ  
__Functions__  
jinspect.MQ_dropDuplicateIDs( )    
  
The next step focuses on improving confidence in the quality of our data. This is done by applying __jinspect.MQ_dropDuplicateIDs( )__ which has the below effects:   
  
* Because one gene can have many proteins, sometimes Maxquant will create multiple proteinGroups for a single gene. As most of our analysis focuses on genes we'll trim the lowest quality proteinGroups duplicates from the table.  
* Standard LFQ defaults require a minimum of 2 peptide species, at least one of which must be unique, for quantitation to be applied. Intensity and iBAQ values, however, do not have such a minimum limit. I consider a 2 peptide minimum to be a wise filter but still have use for the Intensity and iBAQ values. Thus where the LFQ filter is applied all measurements that do not meet the minimum limit will be discarded. In short, if there isn't a companion LFQ value, there won't be an Intensity or iBAQ value either after filtering.  
* It has been documented that Match Between Runs suffers a high frequency of false peptide transfers (Lim, Paulo, Gygi 2019; doi: 10.1021/acs.jproteome.9b00492). At the protein level, however, this false transfer rate is greatly mitigated by the minimum peptide rule applied by the LFQ algorithm. This is another good reason for our filtering step.    

```python
#### Drop duplicates and apply LFQ filter
filter_dict = jinspect.MQ_dropDuplicateIDs(pGroup_clean, metadata, prefix = 'Peptides', ID = 'ENTREZGENE_gPro primary', pool = 'measure', drop_ID = 'None', 
                                            keep_PoolCalcs = False, applyLFQ_filter = ['Intensity', 'iBAQ'])

#### The df_keep value contains our targets, df_droprows conatins the discarded duplicates. Assign the df_keep value to a new variable and inspect.
pGroup_filtered = filter_dict['df_keep']
pGroup_filtered.head(2)
```

### 12. Compare Unique Gene Counts  
__Functions__  
jinspect.MQ_getFrequencyBySample( )  
  
One gene can encode for many proteins that often share regions of similarity. As for illumina-based RNA-Seq, however, shotgun proteomics can rarely assign a peptide species to a singular protein. In MaxQuant these are called proteinGroups. Because we have do not require protein-specific results, and gene identity is more stable, our gene count describes the groups to which our detected proteins have been be assigned. Thus gene here is being detected by protein product, just as it would be detected by RNA product in RNA Seq; none of these 3 are synonymous. To be clear, this is a count and not a measure.

Gene frequency is defined by the summed observations per protein regardless of intensity value and this data is extracted to our modified metadata with __jinspect.MQ_getFrequencyBySample( )__ .   
A typical MQ search will yield identical protein counts (though different values) for Intensity and iBAQ*.  LFQ frequencies will vary depending on the search settings:  
* In this case the MQ search has set LFQ values to be calculated on a min 2 peptide ratio (this is the default)** 

_Notes_

And the same approach as for one or a list of GO ID can be fine tuned for records that have been curated to spcific levels of evidence. The function __jweb.filterQuickGoDict( )__ can be used to filter our records to those of a specific type.  

```python
#### Let's filter our ribozero GO records by experimental evidence and then investigate how many we have recovered.  
dict_ribozero_ExEv = jweb.filterQuickGoDict(dict_ribozero, evidence_code = 'Experimental')

#### Next, annotate the last version of our modified proteinGroups table  
pGroup_ribozero_ExEv = jwrangle.AnnotateDataFrameCtrls(pGroup_Freq, dict_ribozero_ExEv, search_match = 'ENTREZGENE_gPro primary', dict_match = 'ENTREZGENE_gPro primary', none_col = 'GO_None')

#### Next find the counts for each GO ID being searched we will modify our metadata table in a fashion in the same way as for prptide and gene counting.
#### We'll use iBAQ for these counts but, given all intensities have previously been filtered by LFQ membership both 'Intensity' or 'LFQ intensity' would give the same result.
metaStatsRibozero_ExEv = jinspect.MQ_getFrequencyBySample(pGroup_ribozero_ExEv['ann_df'], metaStats, freqList = list(dict_ribozero.keys()) + ['GO_None'], measure = 'iBAQ')
metaStatsRibozero_ExEv

#### Now Calculate the % GO0003723 members per group
metaStatsRibozero_ExEv['% rRNA RBP'] = metaStatsRibozero_ExEv.apply(lambda row: round(100 - (100*row['GO_None']/row['LFQ intensity'])), axis = 1)

pd.set_option('display.max_columns', 500)
metaStatsRibozero_ExEv[metaStatsRibozero_ExEv['condition']=='254-probe']
```

```python
#### First find the number of genes in each category and calculate the % of confirmed genes found
for key, value in dict_ribozero_ExEv.items():
    known = len(value['ENTREZGENE_gPro primary'].unique())
    if known > 0:
        metaStatsRibozero_ExEv['% ' + key + ' (' + str(known) + ')'] = metaStatsRibozero_ExEv.apply(lambda row: round(100*row[key]/known), axis = 1)

#### Relabel the GO IDs with names to make reading easier
MyGO_IDs = dict(zip([i.replace(':','') for i in MyGO_Stats_DF['go ID'].tolist()], MyGO_Stats_DF['GO Term'])) 
col_names = []
for i in list(metaStatsRibozero_ExEv.columns):
    if '% GO' in i:
        for key, value in MyGO_IDs.items():
            if key in i:
                col_names.append(i.replace(key, value))
    else:
        col_names.append(i)
metaStatsRibozero_ExEv.columns = col_names

#### Create Plot for our crosslinked probe samples only (other conditions are not worth doing)
plot = metaStatsRibozero_ExEv[metaStatsRibozero_ExEv['condition']=='254-probe'][['condition'] + [i for i in list(metaStatsRibozero_ExEv.columns) if 'ribosome' in i]]
ribozero_ExEvMelt = pd.melt(plot, "condition", var_name='GO Name (#)')
ribozero_ExEvMelt['GO Name (#)'] = ribozero_ExEvMelt['GO Name (#)'].apply(lambda x: x.replace('% ','')) # we'll remove the % from the names to reduce redundancy from our plot labels.

#### Plot values
bp_bar_edge = dict(linewidth=1.5)
bp_error_edge = dict(capsize=0.2, errcolor=".2", edgecolor=".2",errwidth=1.5)

sns.set(font_scale=1.2)
ax = sns.barplot(x='value', y='GO Name (#)', data=ribozero_ExEvMelt, palette='Reds_d', ci= 'sd', **bp_bar_edge, **bp_error_edge)
ax.set_xlabel('\n% known interactors identified')
```

### Results: NOT BAD!  
Using membership counts to assess the coverage a method can deliver for a specific compartment of molecular function introduces several comprehension issues:  
a) Smaller categories can appear to be of less value than larger categories despite their membership being more restrictive  
b) Categories can vary in size based on their wider research popularity, and this will not reflect physiological importance  
c) Choosing highly permissive levels of evidence, such as computational predictions, clouds judgement of what baseline is achievable during the development of new methods.  
  
Point b) can be addressed by assessing category representation with respect to background lists and Gene Ontology Enrichment Analyses (GOEA). Many online tools are available for doing this and so none are made native to this package. An example of their use can be found in the tutorial for Perturbation Analysis of Dyanmic RBP Networks.

## Conclusion  
The targeted RBP-RNA method has the following positives:  
* Probe first captures greatly outperform the bead capture  
* Fewer RBP are discovered than for global and OdT captures. This suggests that some degree of exclusivity comes from the target hybridisation.  
* Nearly all discovered RBP are RNA-Binding and the nCL samples are very clean.  
* The ribozero kit depends on streptavidin-biotin links and these do not appear to be affected. This is positive news for the use of custom probes.  
* The use of scrambled negative control probes and comparison between different biological conditions would aid in identifying important transcript interactors.  
  
The targeted RBP-RNA method has the following negatives:  
* No scrambled negative control was used in this experiment. It would be difficult to design one because the ribozero probes are of unknown sequence and length.  
* Hybridisation in the presence of trace guanidine, phenol-chloroform, and a high % of formamide means that a thorough investigation of ideal binding temperature is required. This would imply a great deal of effort; ideally using sequencing.
